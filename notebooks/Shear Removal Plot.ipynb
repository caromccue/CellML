```python
from math import sqrt

import numpy as np

import pandas as pd

import matplotlib
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
```

```python
xls = pd.ExcelFile('/Users/carolinemccue/Dropbox (MIT)/Project_ReversibleCellAdhesion_Broad/Shear Experiments/Image Processing/MG63_81922.xlsx')
dfFlat1 = pd.read_excel(xls, 'Flat_1')
dfFlat2 = pd.read_excel(xls, 'Flat_2')
dfFlat3 = pd.read_excel(xls, 'Flat_3')
dfFlat4 = pd.read_excel(xls, 'Flat_4')
df2p5um1 = pd.read_excel(xls, '2.5_1')
df5um1 = pd.read_excel(xls, '5_1')
df10um3 = pd.read_excel(xls, '10_3')
df = pd.read_excel(xls, 'all')


area2p5um1 = df2p5um1['Normalized Area']
area5um1 = df5um1['Normalized Area']
area10um3 = df10um3['Normalized Area']
areaflat1 = dfFlat1['Normalized Area']
areaflat2 = dfFlat2['Normalized Area']
areaflat3 = dfFlat3['Normalized Area']
areaflat4 = dfFlat4['Normalized Area']

shear2p5um1 = df2p5um1['shear (pa)']
shear5um1 = df5um1['shear (pa)']
shear10um3 = df10um3['shear (pa)']
shearflat1 = dfFlat1['shear (pa)']
shearflat2 = dfFlat2['shear (pa)']
shearflat3 = dfFlat3['shear (pa)']
shearflat4 = dfFlat4['shear (pa)']


shear2p5um = df['2.5 um']
shear5um = df['5 um']
shear10um = df['10 um']
shearflat = df['flat']
```

```python
fig = plt.figure(figsize =(10, 7))
ax = fig.add_subplot(111)

plt.plot(shearflat1, areaflat1, '-o', color='k')
plt.plot(shearflat2, areaflat2, '-o', color='k')
plt.plot(shearflat3, areaflat3, '-o', color='k')
plt.plot(shearflat4, areaflat4, '-o', color='k')


#plt.plot(shear2p5um1, area2p5um1, '-o', color='g')
#plt.plot(shear5um1, area5um1, '-o', color='b')
#plt.plot(shear10um3, area10um3, '-o', color='r')


ax.set_xlabel('Shear Stress (Pa)')
ax.set_ylabel('Percent of Cells Adhered')
ax.set_title('Shear Stress to Remove Adhered Cells')
     
# show plot
plt.show()
```

```python
postsize = ['flat', '2.5', '5', '10']
x = np.arange(len(postsize))
y = [shearflat.mean(), shear2p5um.mean(), shear5um.mean(), shear10um.mean()]
errory = [shearflat.std(), shear2p5um.std(), shear5um.std(), shear10um.std()]

fig, ax = plt.subplots()
ax.bar(x, y, yerr = errory, align='center', alpha=0.5, ecolor='black', capsize=10)
ax.set_ylabel('Shear Stress (Pa)')
ax.set_xticks(x)
ax.set_xticklabels(postsize)
ax.set_xlabel('Post Diameter ($\mu m$)')
ax.set_title('Shear Stress to Remove 50% of Adhered Cells')


# Save the figure and show
plt.tight_layout()
plt.show()
```

```python
postsize = ['flat', '2.5', '5', '10']
forceflat = shearflat * 343.440403544253 *0.001
force2p5um = shear2p5um * 122.78701229042194 *0.001
force5um = shear5um * 147.67019913610216 *0.001
force10um = shear10um * 145.75834822805265 *0.001


x = np.arange(len(postsize))
y = [forceflat.mean(), force2p5um.mean(), force5um.mean(), force10um.mean()]
errory = [forceflat.std(), force2p5um.std(), force5um.std(), force10um.std()]

fig, ax = plt.subplots()
ax.bar(x, y, yerr = errory, align='center', alpha=0.5, ecolor='black', capsize=10)
ax.set_ylabel('Shear Force (nN)')
ax.set_xticks(x)
ax.set_xticklabels(postsize)
ax.set_xlabel('Post Diameter ($\mu m$)')
ax.set_title('Shear Force to Remove 50% of Adhered Cells')


# Save the figure and show
plt.tight_layout()
plt.show()
```
