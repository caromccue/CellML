```python
from math import sqrt

import numpy as np

import pandas as pd

import matplotlib
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
```

```python
df = pd.read_excel('/Users/carolinemccue/Dropbox (MIT)/Summer 2022/Shear Experiments/ShearResults.xlsx', sheet_name = 0, header = 0)
shear2p5um = df['2.5 um']
shear5um = df['5 um']
shear10um = df['10 um']
shearflat = df['flat']
```

```python
postsize = ['flat', '2.5', '5', '10']
x = np.arange(len(postsize))
y = [shearflat.mean(), shear2p5um.mean(), shear5um.mean(), shear10um.mean()]
errory = [shearflat.std(), shear2p5um.std(), shear5um.std(), shear10um.std()]

fig, ax = plt.subplots()
ax.bar(x, y, yerr = errory, align='center', alpha=0.5, ecolor='black', capsize=10)
ax.set_ylabel('Shear Force (kPa)')
ax.set_xticks(x)
ax.set_xticklabels(postsize)
ax.set_xlabel('Post Diameter ($\mu m$)')
ax.set_title('Shear Force to Remove 50% of Adhered Cells')


# Save the figure and show
plt.tight_layout()
plt.show()
```

```python
shearflat1 = df['Flat1']
shearflat2 = df['Flat2']
shearflat3 = df['Flat3']
shear = df['Shear (pa)']

fig = plt.figure(figsize =(10, 7))
ax = fig.add_subplot(111)

plt.scatter(shear, shearflat1, color='k')
plt.scatter(shear, shearflat2, color='r')
plt.scatter(shear, shearflat3, color='b')

ax.set_xlabel('Shear Force (kPa)')
ax.set_ylabel('Percent of Cells Adhered')
ax.set_title('Shear Force to Remove Adhered Cells')
     
# show plot
plt.show()
```
