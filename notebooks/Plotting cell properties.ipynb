```python
from math import sqrt

import numpy as np

import pandas as pd

import matplotlib
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
```

```python
%store -r areaumflat
%store -r areaum2p5
%store -r areaum5
%store -r areaum10

%store -r spreading_factorflat
%store -r spreading_factor2p5
%store -r spreading_factor5
%store -r spreading_factor10

%store -r circularityflat
%store -r circularity2p5
%store -r circularity5
%store -r circularity10

%store -r celllengthflat
%store -r celllength2p5
%store -r celllength5
%store -r celllength10

%store -r cellwidthflat
%store -r cellwidth2p5
%store -r cellwidth5
%store -r cellwidth10
```

```python
len(areaumflat), len(areaum2p5), len(areaum5), len(areaum10)
```

```python
postsize = ['flat', '2.5', '5', '10']
x = np.arange(len(postsize))
y = [areaumflat.mean(), areaum2p5.mean(), areaum5.mean(), areaum10.mean()]
errory = [areaumflat.std(), areaum2p5.std(), areaum5.std(), areaum10.std()]

fig, ax = plt.subplots()
ax.bar(x, y, yerr = errory, align='center', alpha=0.5, ecolor='black', capsize=10)
ax.set_ylabel('Cell Area ($\mu m^2$)')
ax.set_xticks(x)
ax.set_xticklabels(postsize)
ax.set_xlabel('Post Diameter ($\mu m$)')
ax.set_title('Total Cell Area on Micropost Surfaces')


# Save the figure and show
plt.tight_layout()
plt.savefig('Cell Area on Microposts')
plt.show()
```
